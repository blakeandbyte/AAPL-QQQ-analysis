## Introduction
In this analysis, we examine Apple Inc. (AAPL) and the Invesco QQQ Trust (QQQ) to understand their historical performance, risk profiles, and potential as investment options. Our goal is to use data-driven insights to inform investment decisions.

## Data Overview
We utilize historical stock data from `yfinance` for AAPL and QQQ, focusing on key metrics like closing prices and volumes from January 1, 2019, to January 1, 2024.

## Fetching and Preparing the Data
First, we gather our data using `yfinance`. This step involves downloading the stock information for both AAPL and QQQ over the specified period.


```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Fetch historical stock data for AAPL and QQQ
tickers = ["AAPL", "QQQ"]
data = yf.download(tickers, start="2019-01-01", end="2024-01-01")['Adj Close']
```

## Exploratory Data Analysis
We begin by examining the dataset to understand its structure and to check for missing values. This step ensures the accuracy of our analysis.

```python
print(data.describe())
print(data.isnull().sum())
```

## Historical Performance Comparison
Comparing the growth trends over time provides insight into the historical performance of AAPL and QQQ. We normalize the prices for a direct comparison.

```python
normalized_data = data / data.iloc[0] * 100
normalized_data.plot(figsize=(14, 7))
plt.title('Normalized Price Trends of AAPL vs QQQ')
plt.ylabel('Normalized Price')
plt.show()
```

## Risk Assessment
Volatility is a key risk indicator. We calculate and compare the standard deviation of daily returns for AAPL and QQQ to assess their risk profiles.


```python
daily_returns = data.pct_change()
volatility = daily_returns.std()
print("Volatility:\n", volatility)

volatility.plot(kind='bar')
plt.title('Volatility Comparison')
plt.show()
```

## Performance Metrics
Analyzing annual returns and Sharpe ratios helps us evaluate the risk-adjusted returns of AAPL and QQQ, providing insight into their investment potential.

```python
annual_returns = daily_returns.mean() * 252
sharpe_ratio = (annual_returns - 0.01) / volatility

print("Annual Returns:\n", annual_returns)
print("Sharpe Ratio:\n", sharpe_ratio)

annual_returns.plot(kind='bar', title='Annual Returns Comparison')
plt.show()

sharpe_ratio.plot(kind='bar', color=['blue', 'green'], title='Sharpe Ratio Comparison')
plt.show()
```
